```python
# to add,subtract,multiply , divide argument  element-wise
import numpy as np
x = np.arange(5)
print("x =", x)
print("x + 3 =", x + 3)      
print("x - 3 =", x - 3)      
print("x * 2 =", x * 2)      
print("x / 2 =", x / 2)       
```

```python
# to round elements of the array to the nearest integer
import numpy as np
x=np.round([9.8])
print("Nearest round integer value",x)
```

```python

import numpy as np 
x=np.array([[3,2],[2,5]])
print("x : \n",x)
print("Mean of x :",np.mean(x))
print("Mean of a along axis=0 :",np.mean(x,axis=0))
print("Mean of a along axis=1 :",np.mean(x,axis=1))
```

```python
#to calculate the diffrence between neighboring element , element wise of a given array
import numpy as np

arr = np.array([1,5,6,45,54,66,6,7,10])
print("Original array:")
print(arr)

ans = np.diff(arr)
print("the diffrence between neighboring element:")
print(ans)
```
