```python
# to convert numpy array to a pandas series
import pandas as pd
import numpy as np
# Creating empty series 
ser = pd.Series() 
print("Pandas Series: ", ser) 
# simple array 
data = np.array(['K','I','N','J','A','L']) 
ser = pd.Series(data) 
print("Pandas Series:\n",ser)
```

```python
#to convert the first column of a Data-frame as series
import pandas as pd  
df = pd.DataFrame() 
print(df)
df = pd.DataFrame({'A':[4,5,6],'B':[7,8,9],}) 
print("Dataframe")
print(df)
```

```python
#to create the mean and standard deviation of the data of a given series
import pandas as pd
s = pd.Series(data = [5, 9, 8, 5, 7, 8, 1, 2, 3, 4, 5, 6, 7, 8, 9, 5, 3])
print(s.mean())
print(s.std())
```

```python
#to sort a given series
import pandas as pd
sr = pd.Series([19.5, 16.8, 22.78, 20.124, 18.1002])
print(sr)
print(sr.sort_values(ascending = True))
```
