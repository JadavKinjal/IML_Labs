```python
#jadav kinjal
#216270307091
#Write a Pandas program to create a line plot of the opening, closing stock prices of given company between two specific dates.
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("/kaggle/input/lab-data/practical_8_data.csv")
start_date = pd.to_datetime('2020-4-1')
end_date = pd.to_datetime('2020-09-30')                         
df['Date'] = pd.to_datetime(df['Date']) 
new_df = (df['Date']>= start_date) & (df['Date']<= end_date)
df2 = df.loc[new_df]
plt.figure(figsize=(12,12))
df2.plot(x='Date', y=['Open', 'Close']);
plt.suptitle('Opening/Closing stock prices of Alphabet Inc.,\n 01-04-2020 to 30-09-2020', fontsize=12, color='black')
plt.xlabel("Date",fontsize=14, color='black')
plt.ylabel("$ price", fontsize=16, color='black')
plt.show()
```
