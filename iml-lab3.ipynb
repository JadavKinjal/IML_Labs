```python
### â€¢ to stack arrays horizontally (column wise)
import numpy as np
i = np.array([ 1, 2, 3] )
print ("1st Input array : \n", i)
in2 = np.array([ 4, 5, 6] )
print ("2nd Input array : \n", in2)
out = np.hstack((i, in2))
print ("Output horizontally stacked array:\n ", out)
 
```
